```python
import cv2
cvim2disp = cv2.imread('zebra.bmp')
cv2.imshow('THIS IS MY ZEBRA', cvim2disp)
cv2.waitKey(0) #image will not show until this is called
cv2.destroyAllWindows() #make sure window closes cleanly
```

```python
figure()
imshow(cvim2disp)
```

What happened?? <br><br>
For historical reasons, OpenCV defaults to USE BGR format instead of usual RGB <br><br>
Solutions: <br><br>
Use numpy/OpenCV consistently<br>
or make sure to convert<br>

```python
#SO CONVERT BGR TO RGB
cvimrgb = cv2.cvtColor(cvim2disp,cv2.COLOR_BGR2RGB) #FIRST IS IMAGE AND SECOND

#or
#imbgr = cv2.cvtColor(cvim2disp,cv2.COLOR_RGB2BGR)

figure()
imshow(cvimrgb)
```

#1 numpy gotcha for people coming from Matlab
#matlab counting from 1 to next, but in python counting from 0 to next

```python
x = zeros(5)
y = x

y[1] - 1

#uncomment next line and run
print(x)
```

What happened?  Why did modifying y change x?
<br><br>A: Python copies arrays and other mutable data types by reference by default
<br><br><br>Here's what you probably want:


```python
x=zeros(5)
y=x.copy()

y[1] = 1

print(x)
```
